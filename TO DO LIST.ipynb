```python
try:
    import Tkinter
    import ttk
except ImportError:  
    import tkinter as Tkinter
    import tkinter.ttk as ttk
    from tkinter import *
    from tkinter import messagebox

windows = Tkinter.Tk()


class MainApp(Tkinter.Frame):
    def __init__(self, windows):
        
        Tkinter.Frame.__init__(self, windows)
        self.windows=windows
        self.infomation_of_task()
        
         
    
    def infomation_of_task(self):
        self.windows.title("My Todolist")
        self.windows.config(background="#B3D9EA")
        self.windows.geometry('1350x700+0+0')
        
        #biáº¿n
        self.task_var = StringVar()
        self.deadline_var = StringVar()
        self.classify_var = StringVar()
        self.status_var = StringVar()
        
        self.title = Tkinter.Label(self.windows, text='My Todolist', font=('Courier new',40, 'bold'), bg='#619196', fg='white')
        self.title.pack(side=TOP, fill=X)
        
        self.info_frame= Tkinter.Frame(self.windows, bd=4, relief=RIDGE, bg='#619196')
        self.info_frame.place(x=20, y=100, width=450, height=500)
        
        self.title_info_lbl= Tkinter.Label(self.info_frame, text='About task', bg='#619196',fg='white', font=('Courier new', 30, 'bold'))
        self.title_info_lbl.grid(row=0, columnspan=2, pady=20)
        
        self.task_lbl=Tkinter.Label(self.info_frame, text='Task', bg='#619196', fg='white', font=('Courier new', 20, 'bold'))
        self.task_lbl.grid(row=1, column=0, pady=10, padx=20, sticky='w')
        
        self.task_ent=Tkinter.Entry(self.info_frame, textvariable=self.task_var,font=('Courier new', 15,'bold'), bd=5, relief=GROOVE)
        self.task_ent.grid(row=1, column=1, pady=10, padx=8, sticky='w')
        
        self.deadline_lbl=Tkinter.Label(self.info_frame, text='Deadline', bg='#619196', fg='white', font=('Courier new', 20, 'bold'))
        self.deadline_lbl.grid(row=2, column=0, pady=10, padx=20, sticky='w')
        
        self.deadline_ent=Tkinter.Entry(self.info_frame, textvariable=self.deadline_var,font=('Courier new', 15,'bold'), bd=5, relief=GROOVE)
        self.deadline_ent.grid(row=2, column=1, pady=10, padx=8, sticky='w')

        self.classify_lbl=Tkinter.Label(self.info_frame, text='Classify', bg='#619196', fg='white', font=('Courier new', 20, 'bold'))
        self.classify_lbl.grid(row=3, column=0, pady=10, padx=20, sticky='w')
        
        self.classify_combo = ttk.Combobox(self.info_frame,textvariable=self.classify_var, font=('Courier new', 13, 'bold'))
        self.classify_combo['values']=('Work', 'Learning', 'Personal')
        self.classify_combo.grid(row=3, column=1, pady=10, padx=20)
        
        self.status_lbl=Tkinter.Label(self.info_frame, text='Status', bg='#619196', fg='white', font=('Courier new', 20, 'bold'))
        self.status_lbl.grid(row=4, column=0, pady=10, padx=20, sticky='w')
        
        self.status_combo = ttk.Combobox(self.info_frame,textvariable=self.status_var, font=('Courier new', 13, 'bold'))
        self.status_combo['values']=('Todo', 'Doing', 'Done')
        self.status_combo.grid(row=4, column=1, pady=10, padx=20)
        
        self.button_frame= Tkinter.Frame(self.info_frame, bd=4, relief=RIDGE, bg='#B3D9EA')
        self.button_frame.place(x=10, y=400, width=430)
        
        self.addbtn=Button(self.button_frame, text='Add', width=10, command=self.add)
        self.addbtn.grid(row=0, column=0, padx=10, pady=10)
        
        
        self.deletebtn=Button(self.button_frame, text='Delete',width=10, command=self.delete)
        self.deletebtn.grid(row=0, column=1, padx=10, pady=10)
        
        self.clearbtn=Button(self.button_frame, text='Clear', width=10, command=self.clear_value)
        self.clearbtn.grid(row=0, column=2, padx=10, pady=10)
        #show task
        self.show_frame= Tkinter.Frame(self.windows, bd=4, relief=RIDGE, bg='#619196')
        self.show_frame.place(x=500, y=100, width=800, height=580)
        
        self.search_ent = Tkinter.Entry(self.show_frame, width=20, font=('Courier new',10, 'bold'), bd=5, relief=GROOVE)
        self.search_ent.grid(row=0, column=2, pady=10, padx=20, sticky='w')
        
        self.searchbtn= Tkinter.Button(self.show_frame, text='Search by task', width=10, pady=5, command=self.search)
        self.searchbtn.grid(row=0, column=3, padx=10, pady=10)
        
      
        
        #show data
        self.table_frame=Tkinter.Frame(self.show_frame, bd=4, relief=RIDGE, bg='#619196')
        self.table_frame.place(x=10, y=60, width=760, height=500)
        
        self.scroll_x= Tkinter.Scrollbar(self.table_frame, orient=HORIZONTAL)
        self.scroll_y= Tkinter.Scrollbar(self.table_frame, orient=VERTICAL)
        self.task_table= ttk.Treeview(self.table_frame, columns=('Task', 'Deadline', 'Classify', 'Status'), xscrollcommand=self.scroll_x.set, yscrollcommand=self.scroll_y.set)
        self.scroll_x.pack(side=BOTTOM, fill =X)
        self.scroll_y.pack(side=RIGHT, fill =Y)
        self.scroll_x.config(command=self.task_table.xview)
        self.scroll_y.config(command=self.task_table.yview)
        self.task_table.heading('Task', text='Task')
        self.task_table.heading('Deadline', text='Deadline')
        self.task_table.heading('Classify', text='Classify')
        self.task_table.heading('Status', text='Status')
        self.task_table['show']='headings'
        self.task_table.column('Task', width=150)
        self.task_table.column('Deadline', width=100)
        self.task_table.column('Classify', width=100)
        self.task_table.column('Status', width=100)
        self.task_table.pack(fill=BOTH, expand=1)
        
    
    # CLEAR TASK
    def clear_value(self):
        self.task_ent.delete(0, END)
        self.deadline_ent.delete(0,END)
        self.status_combo.delete(0, END)
        self.classify_combo.delete(0,END)
     
        
    # ADD TASK
    list_task = []
    def add(self):
        global list_task
        self.tasks = (self.task_ent.get(), self.deadline_ent.get(), self.classify_combo.get(), self.status_combo.get(),self.classify_combo.get() )
        self.list_task.append(self.tasks)
        self.task_table.insert('',END, values = self.tasks )
        messagebox.showinfo('Add task', ' Task added successfully!')
         
    
    list_task = []
    def search(self):
        self.taskOnTreeview = self.task_table.get_children()
        for task in self.taskOnTreeview:
            if self.search_ent.get() in self.task_table.item(task)['values'][0]:
                self.task_search = self.task_table.item(task)['values'][0]
                self.deadline_search = self.task_table.item(task)['values'][1]
                self.status_search = self.task_table.item(task)['values'][3]
                self.classify_search = self.task_table.item(task)['values'][4]
                messagebox.showinfo('Your task', f' Your task is "{self.task_search}". Its classify is "{self.classify_search}". Its status is "{self.status_search}". Its deadline is "{self.deadline_search}"')
            else:
                messagebox.showinfo('Your task', "You don't have this task")
        
    # DELETE TASK
    def delete(self):
        self.selected_task = self.task_table.selection()
        for item in self.selected_task:
            self.task_table.delete(self.selected_task)
            
        
        
def main():
    d=MainApp(windows)
    windows.mainloop()

if __name__=="__main__":
    main()
```
