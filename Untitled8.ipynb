```python
from tkinter import *
from tkinter import ttk
import tkinter as tk

import tkinter.messagebox as messagebox

class Aplication:
    def __init__(self, window):
        self.window = window
        self.window.title("My Todolist")
        self.window.geometry("1350x700+0+0")
        
        title = Lable(self.window, text='My Todolist', font=('times new roman', 40,'bold'), bg='yellow',fg='red')
        title.pack(side=TOP, fill=X)
        
        # các biến 
        self.task_var = StringVar()
        self.deadline_var = StringVar()
        self.classify_var = StringVar()
        self.status_var = StringVar()
        
        
        #khung nhập thông tin nhiệm vụ
        info_frame= Frame(self.window,bd=4, relief=RIDGE, bg='crimson')
        info_frame.place(x=20, y=100, width=450, height=560)
        
        
        title_lbl = Label(info_frame, text='Task infomation',bg='crimson',fg='white', font=('times new roman', 30, 'bold'))
        title_lbl.grid(row=0, columnspan=2, pady=20)
        
        task_lbl = Label(info_frame, text='Task',bg='crimson',fg='white', font=('times new roman', 20, 'bold'))
        task_lbl.grid(row=1, column=0, pady=10, padx=20, sticky='w')
        
        task_ent = Entry(info_frame,textvariable=self.task_var, font=('times new roman', 15, 'bold'), bd=5, relief=GROOVE)
        task_ent.grid(row=1, column=1, pady=10, padx=20, sticky='w')
        
        deadline_lbl = Label(info_frame, text='Deadline',bg='crimson',fg='white', font=('times new roman', 20, 'bold'))
        deadline_lbl.grid(row=1, column=0, pady=10, padx=20, sticky='w')
        
        deadline_ent = Entry(info_frame,textvariable=self.deadline_var, font=('times new roman', 15, 'bold'), bd=5, relief=GROOVE)
        deadline_ent.grid(row=1, column=1, pady=10, padx=20, sticky='w')
        
        classify_lbl = Label(info_frame, text='Classify',bg='crimson',fg='white', font=('times new roman', 20, 'bold'))
        classify_lbl.grid(row=1, column=0, pady=10, padx=20, sticky='w')
        
        combo_classify = ttk.Combobox(info_frame,textvariable=self.classify_var, font=('times new roman', 13, 'bold'))
        combo_classify['values']=('Work', 'School', 'Personal')
        combo_classify.grid(row=4, column=1, padx=20, pady=10)

        
        classify_lbl = Label(info_frame, text='Classify',bg='crimson',fg='white', font=('times new roman', 20, 'bold'))
        classify_lbl.grid(row=1, column=0, pady=10, padx=20, sticky='w')
        
        combo_status = ttk.Combobox(info_frame,textvariable=self.status_var,font=('times new roman', 13, 'bold'))
        combo_status['values']=('Todo', 'Doing', 'Done')
        combo_status.grid(row=5, column=1, padx=20, pady=10)
        
        #khung nút bấm 
        button_frame=Frame(info_frame, bd=4, relief=RIDGE, bg='crimson')
        button_frame.place(x=15,y=500,width=420)
        
        addbtn=Button(button_frame, text="Add", width=10)
        addbtn.grid(row=0, column=0, pads=10, pady=10)
        
        updatebtn=Button(button_frame, text="Update", width=10)
        updatebtn.grid(row=0, column=1, pads=10, pady=10)
        
        deletebtn=Button(button_frame, text="Delete", width=10)
        deletebtn.grid(row=0, column=2, pads=10, pady=10)
        
        clearbtn=Button(button_frame, text="Clear", width=10)
        clearbtn.grid(row=0, column=3, pads=10, pady=10)
        
        #khung chi tiết 
        detail_frame=Frame(self.window,bd=4, relief=RIDGE, bg='crimson')
        detail_frame.place(x=500, y=100, width=800, height=580)
        
        search_lbl = Label(detail_frame, text='Search',bg='crimson',fg='white', font=('times new roman', 20, 'bold'))
        search_lbl.grid(row=0, column=0, pady=10, padx=20, sticky='w')
        
        combo_search = ttk.Combobox(detail_frame,width=10, font=('times new roman', 13, 'bold'))
        combo_search['values']=('Task', 'Deadline', 'Status')
        combo_search.grid(row=5, column=1, padx=20, pady=10)
        
        search_ent = Entry(detail_frame,width=20, font=('times new roman', 10, 'bold'), bd=5, relief=GROOVE)
        search_ent.grid(row=0, column=2, pady=10, padx=20, sticky='w')
        
        searchbtn=Button(button_frame, text="Search", width=10, pady=5)
        searchbtn.grid(row=0, column=3, pads=10, pady=10)
        
        showallbtn=Button(button_frame, text="Show all", width=10, pady=5)
        showallbtn.grid(row=0, column=4, pads=10, pady=10)
       # khung bảng data
        table_frame=Frame(detail_frame,bd=4, relief=RIDGE, bg='crimson')
        table_frame.place(x=10, y=70, width=760, height=500)
        
        scroll_x=Scrollbar(table_frame, orient=HORIZONTAL)
        scroll_y = Scrollbar(table_frame, orient=VERTICAL)
        task_table=ttk.Treeview(table_frame, columns=('Task', 'Deadline', 'classify', 'status'), xscrollcommand=scrollbarx.set, yscrollcommand=scrollbary.set)
        scroll_x.pack(side=BOTTOM, fill=X)
        scroll_y.pack(side=RIGHT, fill=Y)
        scroll_x.config(command=task_table.xview)
        scroll_y.config(command=task_table.yview)
        task_table.heading('Task', text="Task")
        task_table.heading('Deadline', text="Deadline")
        task_table.heading('Classify', text="Classify")
        task_table.heading('Status', text="Status")
        task_table['show']='headings'
        task_table.column('Task', width=150)
        task_table.column('Deadline', width=100)
        task_table.column('Classify', width=100)
        task_table.column('Status', width=100)
        task_table.pack(fill=BOTH, expand=1)
        
window=Tk()
ob=Aplication(window)
window.mainloop()
```
